```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images in Jupyter notebook using matplotlib
def show_image(image, title="Image"):
    plt.figure(figsize=(8, 8))
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the image
image = cv2.imread('problems/rectangle.png')

# Check if the image was successfully loaded
if image is None:
    print("Error: Image not found or unable to read.")
else:
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Draw a vertical line to visualize vertical symmetry
    img_symmetry_vertical = image.copy()
    cv2.line(img_symmetry_vertical, (gray.shape[1] // 2, 0), (gray.shape[1] // 2, gray.shape[0]), (0, 255, 0), 2)

    # Draw a horizontal line to visualize horizontal symmetry
    img_symmetry_horizontal = image.copy()
    cv2.line(img_symmetry_horizontal, (0, gray.shape[0] // 2), (gray.shape[1], gray.shape[0] // 2), (255, 0, 0), 2)

    # Draw diagonal lines to visualize diagonal symmetry (45 and 135 degrees)
    img_symmetry_diagonal = image.copy()
    cv2.line(img_symmetry_diagonal, (0, 0), (gray.shape[1], gray.shape[0]), (0, 0, 255), 2)   # 45 degrees diagonal
    cv2.line(img_symmetry_diagonal, (0, gray.shape[0]), (gray.shape[1], 0), (255, 255, 0), 2) # 135 degrees diagonal

    # Convert images to RGB for matplotlib display
    img_symmetry_vertical_rgb = cv2.cvtColor(img_symmetry_vertical, cv2.COLOR_BGR2RGB)
    img_symmetry_horizontal_rgb = cv2.cvtColor(img_symmetry_horizontal, cv2.COLOR_BGR2RGB)
    img_symmetry_diagonal_rgb = cv2.cvtColor(img_symmetry_diagonal, cv2.COLOR_BGR2RGB)

    # Display the original image and the symmetry visualizations
    show_image(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), title="Original Image")
    show_image(img_symmetry_vertical_rgb, title="Vertical Symmetry Visualization")
    show_image(img_symmetry_horizontal_rgb, title="Horizontal Symmetry Visualization")
    show_image(img_symmetry_diagonal_rgb, title="Diagonal Symmetry Visualization (45° and 135°)")
```
